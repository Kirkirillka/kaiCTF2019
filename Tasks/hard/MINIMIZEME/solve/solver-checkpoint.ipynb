# 4. Online Optimizer

```python
import requests
import sklearn
import numpy as np
import pandas as pd
from scipy.optimize import least_squares,minimize,fmin_cg, basinhopping, minimize_scalar



diff_link = "http://127.0.0.1:10000/diff/{0:.12f}/{0:.12f}"
flag_link = "http://127.0.0.1:10000/try/{0:.8f}/{0:.8f}"

def get_cost_evaluation(x_,y_):

    actual_link= diff_link.format(x_,y_)
    #print(actual_link)
    ans = requests.get(actual_link).json()

    return ans['distance']

def get_flag_for_position(x_,y_):

    actual_link = flag_link.format(x_, y_)
    #print(actual_link)
    ans = requests.get(actual_link).json()

    return ans

def get_c_eval_wrapper(x):
    cost = get_cost_evaluation(x[0],x[1])
    print(x, cost)
    return cost
```

```python
# Random point to start
eta0 = np.array([np.random.random(),np.random.random() ])

#pos = minimize(get_c_eval_wrapper, eta0, method="CG",options={"gtol": 1e-4,"maxiter":50})

pos = minimize(get_c_eval_wrapper, eta0,method='Nelder-Mead', tol=1e-2, options={"maxiter": 50})
```
